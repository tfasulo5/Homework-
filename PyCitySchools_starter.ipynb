```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_pd = "Resources/schools_complete.csv"
student_data_pd= "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_pd = pd.read_csv(school_data_pd)
student_data_pd = pd.read_csv(student_data_pd)

# Combine the data into a single dataset
new_school_data_pd = pd.merge(student_data_pd, school_data_pd, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
new_school_data_pd.columns
new_school_data_pd.head()
```

```python
schools_count = new_school_data_pd["school_name"].nunique()
number_of_students = new_school_data_pd["Student ID"].nunique()

total_budget = school_data_pd["budget"].sum()
average_math_score = new_school_data_pd["math_score"].mean()
average_reading_score = new_school_data_pd["reading_score"].mean()


new_school_data_pd["Passing"] =  (new_school_data_pd["math_score"] + new_school_data_pd["reading_score"] ) / 2
new_school_data_pd["Passing"] 



math_greater_than_70 = new_school_data_pd.loc[new_school_data_pd["math_score"] >= 70, ["math_score"]].count()
math_percent = math_greater_than_70/number_of_students
#print(math_percent.iloc[0])


reading_greater_than_70 = new_school_data_pd.loc[new_school_data_pd["reading_score"] >= 70, ["reading_score"]].count()
reading_percent = reading_greater_than_70/number_of_students
#print(reading_percent.iloc[0])

overall = (math_percent.iloc[0] + reading_percent.iloc[0]) / 2
#print(overall)

summary_pd = pd.DataFrame({
      "number_of_schools":[schools_count],
      "total_students":[number_of_students],
      "sum_of_budget":[total_budget],
      "math_score_average":[average_math_score],
      "reading_score_average":[average_reading_score],
      "percent_math":math_percent.iloc[0],
      "percent_reading":reading_percent.iloc[0],
      "overall_passing_rate":[overall]
})


summary_pd.style.format({                       
                         "sum_of_budget": "${:.0f}",
                         "math_score_average": "{:.1f}%",
                         "reading_score_average": "{:.1f}%",
                         "percent_math": "{:.1%}",
                         "percent_reading": "{:.1%}",
                         "overall_passing_rate": "{:.1%}"
})




    
```

```python
schools_names = school_data_pd.set_index("school_name") ["type"]

all_students = new_school_data_pd["school_name"].value_counts() 

total_school_budget = school_data_pd.set_index("school_name")["budget"]

per_school = school_data_pd.set_index("school_name") ["budget"] / school_data_pd.set_index("school_name")["size"]

average_math_score = student_data_pd.groupby("school_name")["math_score"].mean()

average_reading_score = student_data_pd.groupby("school_name")["reading_score"].mean()

percent_math = student_data_pd[student_data_pd["math_score"]>=70].groupby("school_name")["Student ID"].count() / all_students

percent_reading = student_data_pd[student_data_pd["reading_score"]>=70].groupby("school_name")["Student ID"].count() / all_students

Passing =  (percent_math + percent_reading) /2

school_pd = pd.DataFrame({
    "school_type":schools_names,
    "students":all_students,
    "total_budget":total_school_budget,
    "per_student":per_school,
    "math_score_average":average_math_score,
    "reading_score_average":average_reading_score,
    "math_percents":percent_math,
    "reading_percents":percent_reading,
    "average_passing_rate":Passing
    
      })
school_pd





```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
    
school_pd["school_bins"] = pd.cut(school_pd["per_student"], spending_bins, labels=group_names)
school_pd["school_bins"]

bin_school = school_pd.groupby("school_bins")
ave_math = bin_school.mean()["math_score_average"]
ave_math
 
bin_school = school_pd.groupby("school_bins")
ave_reading = bin_school.mean()["reading_score_average"]
ave_reading

bin_school = school_pd.groupby("school_bins")
math_per = bin_school.mean()["math_percents"]
math_per

bin_school = school_pd.groupby("school_bins")
reading_per = bin_school.mean()["reading_percents"]
reading_per

bin_school = school_pd.groupby("school_bins")
overall_per = bin_school.mean()["average_passing_rate"]
overall_per

bins_budget = pd.DataFrame({
    "Math_Average": ave_math,
    "Reading_Average": ave_reading,
    "Percent_Math": math_per,
    "Percent_Reading":reading_per,
    "Percent_Pass": overall_per
    
})
bins_budget
```
